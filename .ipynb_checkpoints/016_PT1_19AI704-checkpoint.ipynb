```python
import pandas as pd

url = "https://raw.githubusercontent.com/toshihiroryuu/Dataset-test/main/Wine.csv"
# read wine data set. 
df = pd.read_csv(url)

df.head()
```

```python
# identify the size of data frame
print("Size of the data frame is",df.size)

# identify the shape of the data frame 
print("Shape of the data frame is",df.shape)
print("No of Rows = ", df.shape[0])
print("No of Columns = ", df.shape[1])

df.describe()
```

```python
pd.DataFrame({'mean': df.mean(),
'sd': df.std(),
'min': df.min(),
'max': df.max(),
'median': df.median(),
'length': len(df),
'miss.val': df.isnull().sum(),
})
```

```python
df = df.drop(['Type'], axis=1)
df.head(5)
```

### Handle missing values

```python
miss_count = df.isnull().sum().sum()

print("No of missing values is",miss_count)
```

No missing values in the dataset.

### Normalise data

```python
from sklearn import preprocessing

np_norm = preprocessing.scale(df)
df_norm = pd.DataFrame(np_norm, columns = df.columns)

print("Shape of the data frame is",df_norm.shape)
df_norm.head(5)
```

### PCA

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing

pcs = PCA()
pcs.fit(df_norm)
```

```python
import numpy as np

pcsSummary_df = pd.DataFrame({"Std devn":np.sqrt(pcs.explained_variance_),
                              "Proportion of var": pcs.explained_variance_ratio_,
                              "Cum proportion": np.cumsum(pcs.explained_variance_ratio_)})
```

```python
pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ["PC"+str(i) for i in range(1,len(pcsSummary_df.columns) + 1)]
```

```python
pcsSummary_df.round(4)
```

    Principal Components 1 and 2 (PC1 & PC2) contains 55% of the information cumulatively.Principal Components 1 and 2 (PC1 & PC2) contains and 36% and 19.2 % of the information individually.

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(),
                                columns=pcsSummary_df.columns,
                                index=df.columns)

pcsComponents_df
```
