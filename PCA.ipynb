```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/toshihiroryuu/Dataset-test/main/Cereals.csv")
df
```

```python
pcs = PCA(n_components=2)
pcs.fit(df[["calories", "rating"]])
```

```python
pcsSummary = pd.DataFrame({"Standard deviation":np.sqrt(pcs.explained_variance_),"Proportion of variance": pcs.explained_variance_ratio_,
"Cumulative proportion": np.cumsum(pcs.explained_variance_ratio_)})
```

```python
pcsSummary
```

```python
pcsSummary= pcsSummary.transpose()
pcsSummary
```

```python
pcsSummary.columns = ["PC1", "PC2"]
pcsSummary.columns
```

```python
pcsSummary.round(4)
```

```python
pcs.components_
```

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), columns=["PC1","PC2"],index=["calories", "rating"])
pcsComponents_df
```

```python
scores = pd.DataFrame(pcs.transform(df[["calories", "rating"]]),columns=["PC1", "PC2"])
scores
```
